### Seaborn

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
#  plot the distribution of the data ---> like a histogram
```

```python
sns.displot(iris['petal_length'],bins=20,kde=True)
```

```python
sns.kdeplot(iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'],y=iris['petal_width'])
plt.show()
```

```python
sns.scatterplot(x=iris['petal_length'],y=iris['petal_width'],data=iris,hue='species')
plt.show()
```

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,kind='reg')
```

### pair plot

```python
sns.pairplot(data=iris,hue='species')
```

## Categorical plots

```python
sns.countplot(x='species',data=iris)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='species',y="sepal_length",data=iris)
plt.show()
```

```python
sns.violinplot(x='species',y='petal_length',data=iris)
```

## finding correlation

```python
iris.corr()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_length',data=iris)
```

```python
sns.heatmap(iris.corr(),annot=True,cmap="Blues")
```

## Visualize tips dataset

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
sns.countplot(x=tips['sex'])
```

```python
tips['day'].unique()
```

```python
# out of all the days which day gets the highest amount
# Ans-Sunday

sns.barplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# what do people prefer-lunch or dinner ?
sns.barplot(x='time',y='total_bill',data=tips,hue='day')
```

```python
# do smoker people give more tip?
sns.barplot(x='smoker',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```
