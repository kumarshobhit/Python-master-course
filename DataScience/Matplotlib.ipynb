movie visulization

```python
movie=pd.read_csv('./movie_meta.csv')
movie.head()
```

```python
movie.columns
```

```python
movie["movie_title"]
```

```python
movie["movie_title"][0]
```

```python
def preprocess(ele):
    return len(ele.strip())
```

```python
movie["title_length"]=movie["movie_title"].apply(preprocess)
```

```python
plt.hist(movie["title_length"],bins=50)
plt.xticks(range(0,100,5))
plt.xlabel("length of movie titles")
plt.ylabel("frequency")
plt.show()
```

```python
len_dict={}
for i in movie["title_length"]:
    if len_dict.get(i) is None:
        len_dict[i]=1 
    else :
        len_dict[i]+=1
        
            
```

```python
len_dict.keys()
len_dict.values()
```

```python
x=len_dict.keys()
y=len_dict.values()
plt.scatter(x,y)
plt.xticks(range(0,60,5))
plt.xlabel("Movies title length")
plt.ylabel("Number of movies")
plt.show()
```
